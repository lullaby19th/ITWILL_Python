## 변경된 점심시간 문제 : 표본 오차란?

링크 : https://hongsi1991.tistory.com/83  ( 참고 : 히스토그램 vs 막대그래프)

모집단에서 복수의 표본을 무작위로 추출하면 곤란한 문제가 생긴다.    
모집단의 성질이 표본별로 달라져 버리는데 이 성질을 표본오차라 한다.  
이 표본오차를 극복하는 것이 확률론이다.  

__모집단__ : 십만명의 인구중 20%의 흡연자가 있는 집단.

```python
# -*- coding: utf-8 -*-
"""
Created on Mon Dec  9 11:03:34 2019

@author: wdp
"""

import random
from random import shuffle

a = []
people = int(input('전체 인구수를 입력하세요.'))

for i in range(int(people*0.2)): #흡연자를 1로가정
    a.append(1)

for i in range(int(people*0.8)): # 비흡연자를 0으로 가정
    a.append(0)


shuffle(a)


size = int(input('100명의 표본을 몇번 추출할것인지 입력하시오!'))

for i in range(1,size+1):
    b = random.sample(a,100)
    c = (sum(b)/100) * 100

    print('포본', i, '의 흡연률 :', c, '%')
```

## 문제240.   
## 금요일에 냈던 마지막 문제인 중심극한의 정리의 결과를 d드라이브 밑에  
## mydata2.txt라는 이름으로 생성하시오!  

```python
def dice_average(cnt):
    
    import random
    import numpy as np

    dice = 100000

    c = []

    for i in range(cnt):

        a = [1,2,3,4,5,6] * dice

        b = random.sample(a, 10)

        c.append(np.mean(b))
        d = np.mean(c)

    return d

cnt = [10,50,100,500,1000]

result = map(dice_average, cnt)

f = open("d:\\data\\mydata2.txt", "w")

f.write(str(list(result))) #결과값이 map이라서 앞에 list로 감싸준다.

f.close()
```

## 예제137. 텍스트 파일에 한 줄씩 쓰기 (writelines)

파일 객체의 writelines()는 텍스트 문자열이나 텍스트 문자열이 요소로 되어있는 리스트를 인자로 받아  
파일에 한 줄씩 기록합니다.  
리스트가 인자인 경우 writelines()는 리스트의 요소를 하나의 문자열로 결합하여 파일에 한꺼번에 기록합니다.

★ writelines 는 리스트 자료형도 파일에 저장할 수 있다.

```python
listdata = ['a', 'b', 'c', 'd', 'e']
result1 = sorted(listdata)

print(result1)

f = open("d:\\data\\mydata4.txt", "w")
f.writelines(result1) #리스트 안의 요소가 문자일시 write를 하면 오류가남
f.close()
```

### ※ 위와 같이 리스트 안의 요소가 숫자가 아니라 문자이면 write로 하면 에러가 나고 writelines로 하면 수행된다.

## 예제139. 바이너리 파일 복사하기(read, write)

일반적으로 바이너리 파일은 용량이 큰 경우가 많다.  
예를들어, 고화질 영화를 담은 동영상 파일의 경우 10GB가 넘기도 한다.  
대용량 파일의 내용을 한 번에 읽고 쓰는 것은 메모리 용량의 한계로 불가능할 때가 있고 성능적인 면에서도 효율적이지 못하다.  
이러한 경우 파일을 처음부터 끝까지 일정한 크기 단위로 읽고 쓰면 메모리 문제를 해결할 수 있다.

```python
bufsize = 1024 # 1kb = 1024byte

f = open("d:\\data\\lena.png", "rb")
h = open("d:\\data\\lena_copy.png", "wb")

data = f.read(bufsize) # 1kb씩 읽어오는 상황

while data:
    h.write(data)
    data = f.read(bufsize) # 파일이 너무 크기때문에 while반복문을 이용하여 1kb씩 읽고 복사하는(쓰는) 상황.
    
f.close()
h.close()
```

```python
bufsize = 1024*5 # 1kb = 1024byte

f = open("d:\\data\\lena.png", "rb")
h = open("d:\\data\\lena_copy2.png", "wb")

data = f.read(bufsize) # 1kb씩 읽어오는 상황
while data:
    h.write(data)
    data = f.read(bufsize) # 파일이 너무 크기때문에 while반복문을 이용하여 1kb씩 읽고 복사하는(쓰는) 상황.
f.close()
h.close()
```

## 문제253. a.html 문서에서 a태그의 href 링크의 url만 긁어 오시오!

```python
from bs4 import BeautifulSoup

with open("d:\\data\\a.html", encoding = "UTF-8") as a:
    soup = BeautifulSoup(a, "lxml") # 이렇게 하면 soup에 html 전체가 soup에 담긴다.

for i in soup.find_all('a'):
    print(i) # 리스트 형이기 때문에 하나하나씩 for문으로 가져와야된다.
    print(i.get("href")) # i.get("href")로 url만 깔끔하게 긁어올 수 있다.
```
